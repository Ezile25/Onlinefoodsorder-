```python
onlinefoods_dataset = pd.read_csv("onlinefoods.csv")
```

```python
#Statistics for Age
age_mean = onlinefoods_dataset['Age'].mean()
age_std_error = onlinefoods_dataset['Age'].sem()
age_median = onlinefoods_dataset['Age'].median()
age_mode = onlinefoods_dataset['Age'].mode()[0]
age_std_dev = onlinefoods_dataset['Age'].std()
age_variance = onlinefoods_dataset['Age'].var()
age_kurtosis = kurtosis(onlinefoods_dataset['Age'], fisher=False)
age_skewness = skew(onlinefoods_dataset['Age'])
age_range = onlinefoods_dataset['Age'].max() - onlinefoods_dataset['Age'].min()
age_min = onlinefoods_dataset['Age'].min()
age_max = onlinefoods_dataset['Age'].max()
age_total = onlinefoods_dataset['Age'].sum()
age_count = onlinefoods_dataset['Age'].count()
age_percentiles = np.percentile(onlinefoods_dataset['Age'], [25, 50, 75, 90, 95, 99])
```

```python
#statistics for family size
family_mean = onlinefoods_dataset['Family size'].mean()
family_std_error = onlinefoods_dataset ['Family size'].sem()
family_median = onlinefoods_dataset ['Family size'].median()
family_mode = onlinefoods_dataset['Family size'].mode()[0]
family_std_dev = onlinefoods_dataset['Family size'].std()
family_variance = onlinefoods_dataset ['Family size'].var()
family_kurtosis = kurtosis(onlinefoods_dataset ['Family size'], fisher=False)
family_skewness = skew(onlinefoods_dataset ['Family size'])
family_range = onlinefoods_dataset ['Family size'].max() - onlinefoods_dataset ['Family size'].min()
family_min = onlinefoods_dataset ['Family size'].min()
family_max = onlinefoods_dataset ['Family size'].max()
family_total = onlinefoods_dataset ['Family size'].sum()
family_count = onlinefoods_dataset ['Family size'].count()
family_percentiles = np.percentile(onlinefoods_dataset['Family size'], [25, 50, 75, 90, 95, 99])

```

```python
#statistics for latitude
latitude_mean = onlinefoods_dataset ['latitude'].mean()
latitude_std_error = onlinefoods_dataset ['latitude'].sem()
latitude_median = onlinefoods_dataset ['latitude'].median()
latitude_mode = onlinefoods_dataset ['latitude'].mode()[0]
latitude_std_dev = onlinefoods_dataset ['latitude'].std()
latitude_variance = onlinefoods_dataset ['latitude'].var()
latitude_kurtosis = kurtosis(onlinefoods_dataset ['latitude'], fisher=False)
latitude_skewness = skew(onlinefoods_dataset ['latitude'])
latitude_range = onlinefoods_dataset ['latitude'].max() - onlinefoods_dataset['latitude'].min()
latitude_min = onlinefoods_dataset ['latitude'].min()
latitude_max = onlinefoods_dataset ['latitude'].max()
latitude_total = onlinefoods_dataset ['latitude'].sum()
latitude_count = onlinefoods_dataset ['latitude'].count()
latitude_percentiles = np.percentile(onlinefoods_dataset ['latitude'], [25, 50, 75, 90, 95, 99])
```

```python
#statistics for longitude
longitude_mean = onlinefoods_dataset ['longitude'].mean()
longitude_std_error = onlinefoods_dataset['longitude'].sem()
longitude_median = onlinefoods_dataset ['longitude'].median()
longitude_mode = onlinefoods_dataset ['longitude'].mode()[0]
longitude_std_dev = onlinefoods_dataset ['longitude'].std()
longitude_variance = onlinefoods_dataset ['longitude'].var()
longitude_kurtosis = kurtosis (onlinefoods_dataset ['longitude'], fisher=False)
longitude_skewness = skew (onlinefoods_dataset['longitude'])
longitude_range = onlinefoods_dataset['longitude'].max() - onlinefoods_dataset ['longitude'].min()
longitude_min = onlinefoods_dataset ['longitude'].min()
longitude_max = onlinefoods_dataset ['longitude'].max()
longitude_total = onlinefoods_dataset ['longitude'].sum()
longitude_count = onlinefoods_dataset ['longitude'].count()
longitude_percentiles = np.percentile(onlinefoods_dataset ['longitude'], [25, 50, 75, 90, 95, 99])
```

```python
#statistics for pincode
#converting non numeric values to numeric
onlinefoods_dataset['Pin code'] = pd.to_numeric(onlinefoods_dataset['Pin code'], errors='coerce')
#dropping nan values
onlinefoods_dataset.dropna(subset=['Pin code'], inplace=True)

pincode_mean = onlinefoods_dataset['Pin code'].mean()
pincode_std_error = onlinefoods_dataset['Pin code'].sem()
pincode_median = onlinefoods_dataset['Pin code'].median()
pincode_mode = onlinefoods_dataset['Pin code'].mode()[0]
pincode_std_dev = onlinefoods_dataset['Pin code'].std()
pincode_variance = onlinefoods_dataset['Pin code'].var()
pincode_kurtosis = kurtosis(onlinefoods_dataset['Pin code'], fisher=False)
pincode_skewness = skew(onlinefoods_dataset['Pin code'])
pincode_range = onlinefoods_dataset['Pin code'].max() - onlinefoods_dataset['Pin code'].min()
pincode_min = onlinefoods_dataset['Pin code'].min()
pincode_max = onlinefoods_dataset['Pin code'].max()
pincode_total = onlinefoods_dataset['Pin code'].sum()
pincode_count = onlinefoods_dataset['Pin code'].count()
pincode_percentiles = np.percentile(onlinefoods_dataset['Pin code'], [25, 50, 75, 90, 95, 99])



```

```python
#Destriptive statistics for onlinefoods dataset
descriptive_stats = {}

features = ['Age', 'Family size', 'latitude', 'longitude', 'Pin code']


for feature in features:
    mean = onlinefoods_dataset[feature].mean()
    std_error = onlinefoods_dataset[feature].sem()
    median = onlinefoods_dataset[feature].median()
    mode = onlinefoods_dataset[feature].mode()[0]
    std_dev = onlinefoods_dataset[feature].std()
    variance = onlinefoods_dataset[feature].var()
    kurt = kurtosis(onlinefoods_dataset[feature], fisher=False)
    skewness = skew(onlinefoods_dataset[feature])
    rng = onlinefoods_dataset[feature].max() - onlinefoods_dataset[feature].min()
    min_val = onlinefoods_dataset[feature].min()
    max_val = onlinefoods_dataset[feature].max()
    total = onlinefoods_dataset[feature].sum()
    count = onlinefoods_dataset[feature].count()
    percentiles = np.percentile(onlinefoods_dataset[feature], [25, 50, 75, 90, 95, 99])

    
    descriptive_stats[feature] = {
        'Mean': mean,
        'Standard Error': std_error,
        'Median': median,
        'Mode': mode,
        'Standard Deviation': std_dev,
        'Variance': variance,
        'Kurtosis': kurt,
        'Skewness': skewness,
        'Range': rng,
        'Minimum': min_val,
        'Maximum': max_val,
        'Sum': total,
        'Count': count,
        'Percentiles': percentiles
    }

descriptive_stats_df = pd.DataFrame(descriptive_stats)

descriptive_stats_df = descriptive_stats_df.T

print(descriptive_stats_df)


```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(onlinefoods_dataset['Age'], 
            notch=True,
            vert=False,              
            patch_artist=True,       
            boxprops=dict(facecolor='skyblue', color='black'),
            whiskerprops=dict(color='black', linestyle='--'),   
            capprops=dict(color='black'),                       
            medianprops=dict(color='red', linewidth=2),        
            flierprops=dict(marker='o', markersize=8, markerfacecolor='green', markeredgecolor='black'))  # Outlier style

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Values')
plt.title('Box and Whisker Plot for Age')

# Display the plot
plt.show()
```

```python
# Bar chart of the 'Gender' column
gender_counts = onlinefoods_dataset['Gender'].value_counts()
plt.figure(figsize=(6, 6), dpi=100)  
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='Set2')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gender')
for index, value in enumerate(gender_counts.values):
    plt.text(index, value, str(value), ha='center', va='bottom')
plt.show()
```
